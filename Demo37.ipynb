```python
import pandas as pd
data = pd.read_csv("/content/global-malaria-deaths/global-malaria.csv")
data.head() # Preview the data
```

```python
import pandas as pd
# Load the dataset (correct path)
df = pd.read_csv("/content/global-malaria-deaths/global-malaria.csv")
# Verify
print(df.head())
print("\nColumns:", df.columns.tolist())
```

```python
data['Region'].value_counts().plot(kind='bar', title='Region Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
top_regions = data['Region'].value_counts().head(10)  # Get top 10 regions
top_regions.plot(kind='bar', title='Top 10 Regions by Count')
plt.xticks(rotation=45, ha='right')  # Rotate labels for better readability
plt.tight_layout()  # Prevent label cutoff
plt.show()
```

```python
plt.figure(figsize=(10, 5))  # Wider figure
data['Region'].value_counts().head(10).plot(kind='barh')  # Horizontal bars
plt.title('Top 10 Regions by Count', pad=20)
plt.tight_layout()
plt.show()
```

```python
# Melt the data to long format for better visualization
melted_data = data.melt(id_vars=['Region'],
                       var_name='Year',
                       value_name='Deaths')

# Convert Year to numeric
melted_data['Year'] = pd.to_numeric(melted_data['Year'])

# Plot line chart for each region
for region in melted_data['Region'].unique():
    region_data = melted_data[melted_data['Region'] == region]
    plt.plot(region_data['Year'], region_data['Deaths'], label=region)

plt.title('Malaria Deaths Over Time by Region')
plt.xlabel('Year')
plt.ylabel('Deaths')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Select a year column (e.g., 2019) to plot
data['2019'].plot(kind='hist', bins=20, title='Distribution of Malaria Deaths in 2019')
plt.xlabel('Deaths')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter Afghanistan's data (correct spelling)
afghanistan_data = data[data['Region'] == 'Afghanistan'].copy()

# Reshape data: Years → Rows
afghanistan_yearly = afghanistan_data.drop('Region', axis=1).T.reset_index()
afghanistan_yearly.columns = ['Year', 'Deaths']  # Rename columns
afghanistan_yearly['Year'] = pd.to_numeric(afghanistan_yearly['Year'])  # Ensure 'Year' is numeric
```

```python
plt.figure(figsize=(12, 6))
sns.regplot(x='Year', y='Deaths', data=afghanistan_yearly,
            scatter_kws={'color': 'red'},
            line_kws={'color': 'blue', 'linestyle': '--'},
            lowess=True)  # Smooth trendline
plt.title('Malaria Death Trend in Afghanistan (1990–2019)')
plt.grid(True)
plt.show()
```
